```python
from pandas.io import data as web
```

```python
all_data={}
```

```python
for ticker in ['AAPL','IBM','MSFT','GOOG']:
    all_data[ticker]=web.get_data_yahoo(ticker,'1/1/2000','1/1/2010')
```

```python
all_data #dict object
```

```python
all_data['IBM'] #DataFrame Object 
```

```python
price=DataFrame({tic:data['Adj Close'] for tic,data in all_data.items()}) #字典生成式
```

```python
volumn=DataFrame({tic:data['Volume'] for tic,data in all_data.items()}) #字典生成式
```

```python
price
```

```python
volumn
```

```python
#compute percent changes of the prices, DataFrame.pct_change() method
```

```python
returns=price.pct_change()
```

```python
returns.tail() #DataFrame.tail(n=5), Returns last n rows
```

```python
#Series.corr() method, Series.cov() method
#The corr method of Series computes the correlation of the overlapping, non-NA,
#aligned-by-index values in two Series. Relatedly, cov computes the covariance
```

```python
returns.MSFT.corr(returns.IBM) #returns-->DataFrame, returns.MSFT-->Series,
                               #Series.corr(), 計算兩個Series的相關係數
```

```python
returns.MSFT.cov(returns.IBM) #Series.cov(), 計算兩個Series的協方差(covariance) 
```

```python
#DataFrame’s corr and cov methods, on the other hand, return a full correlation 
#or covariance matrix as a DataFrame, respectively
```

```python
returns.corr()
```

```python
returns.cov()
```

```python
#DataFrame.corrwith(), method
#引數如果是Series --> 返回一個相關係數值
#引述如果是DataFrame --> 則會計算按欄配對的相關係數 
```

```python
returns.corrwith(returns.IBM) #retuens.IBM --> Series object
```

```python
returns.corrwith(volumn) #volumn -->DataFrame objecdt
```
