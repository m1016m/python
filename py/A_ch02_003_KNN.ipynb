```python
#產生第一個資料集, mglearn.datasets.make_forge()
import mglearn
x,y=mglearn.datasets.make_forge()
```

```python
#檢視 make_forge()產生的訓練資料集, y為label值: 0-->'o', 1-->'^'
print('len(x):',len(x),' x.shape:',x.shape,' y.shape:',y.shape,' x[0]:',x[0],' y[0]:',y[0])
```

```python
#在這個最簡版本中,只考慮唯一一個最近的臨近點
import matplotlib.pyplot as plt
%matplotlib inline
#以mglearn.discrete_scatter()繪圖檢視訓練資料集
mglearn.discrete_scatter(x[:,0],x[:,1],y)
```

```python
#k=1, one-nearest-neighbor, 本範例只是就結果繪圖說明
#三個測試點
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
#指定k為任意數值(這是k-NN命名的由來),用投票來決定屬於哪種label
```

```python
#k=3 影响的人变多了 和所有人计算 找出离我最近的三个人 判断其中最近的那个人
mglearn.plots.plot_knn_classification(n_neighbors=5)
```
