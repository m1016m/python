```python
#Plots in matplotlib reside within a Figure object.
#You can create a new figure with plt.figure
fig=plt.figure()
```

```python
#You canâ€™t make a plot with a blank figure. 
#You have to create one or more subplots using add_subplot:
fig2=plt.figure()
ax1=fig2.add_subplot(2,2,1)  #2x2å€‹subplotä¸­çš„ç¬¬ä¸€å€‹
ax2=fig2.add_subplot(2,2,2)  #2x2å€‹subplotä¸­çš„ç¬¬äºŒå€‹
ax3=fig2.add_subplot(2,2,3)  #2x2å€‹subplotä¸­çš„ç¬¬ä¸‰å€‹
plt.plot(np.random.randn(50).cumsum(), 'k--')  #ç•«åœ¨æœ€å¾Œä¸€å€‹, ax3
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)  #ç•«åœ¨ ax1, hist
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30)) #ç•«åœ¨ ax2,scatter
```

```python
#plt.subplots(), is a convenience method that creates a new figure and 
#returns a NumPy array containing the created subplot objects
```

```python
fig2, subplt = plt.subplots(2,3) #ä»¥plt.subplot() å»ºç«‹ä¸€å€‹ figure object, 'fig2', å…§å« 2x3 å€‹subplot ç‰©ä»¶
subplt[1,1].plot(np.random.rand(500)) #åœ¨ç¬¬ [1,1] å€‹ subplot, subplt[1,1] ä¸Šç•«åœ–
```

```python
#subplot, å…±ç”¨åº§æ¨™--> sharex=True, sharey=True
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='g', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```
