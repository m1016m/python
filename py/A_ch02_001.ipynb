```python
#產生第一個資料集, mglearn.datasets.make_forge()
import mglearn
x,y=mglearn.datasets.make_forge()
```

```python
mglearn.datasets.make_forge?
```

```python
#檢視 make_forge() 產生的合成資料
print('x.shape-->',x.shape,' y.shapr-->',y.shape,' x[0]-->',x[0],' y[0]-->',y[0])
```

```python
#繪製 make_forge() 產生的合成資料資料集
import matplotlib.pyplot as plt
%matplotlib notebook
mglearn.discrete_scatter(x[:,0],x[:,1],y)
plt.legend(["Class 0","Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
```

```python
#產生第二個資料集(波型資料), mglearn.datasets.make_wave()
x,y=mglearn.datasets.make_wave(n_samples=40)
```

```python
#檢視 make_wave() 產生的資料
print('x.shape-->',x.shape,' y.shape-->',y.shape,' x[0]-->',x[0],' y[0]-->',y[0])
```

```python
#將 make_wave() 產生的資料繪製圖型
plt.plot(x,y,'ro')    #r--> red, o--> 圈圈, 你可以試試 'g^' 
plt.ylim(-3,3)
plt.xlabel("Feature")
plt.ylabel("Target")
```
