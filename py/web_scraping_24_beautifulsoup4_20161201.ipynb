```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url='http://www.timeanddate.com/weather/'
```

```python
html=requests.get(url).text
```

```python
bsObj=BeautifulSoup(html,"html.parser") #以BeautifulSoup() 將html轉成 Beautifulsoup object
```

```python
print(bsObj.prettify()) #display bsObj
```

```python
links=bsObj.find_all('a') #以find_all() method,找出所有 tag <a>, links--> list of tag <a>
```

```python
links[0:4] #list中前三個元素
```

```python
links[10]
```

```python
links[10].contents
```

```python
links[10].get_text() #get_text() method
```

```python
links[10].get('href') #取得tag中 href屬性值
```
