```python
#Two_moons dataset 產生說明
import mglearn
from sklearn.datasets import make_moons #sklearn 的 make_moons() function
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#畫資料點, plt.scatter()
X,y=make_moons(n_samples=100, noise=0.25, random_state=3)
X1 =[X[i][0] for i in range(100)]
X2 =[X[i][1] for i in range(100)]
colors = (0,0,0)
area = np.pi*3
plt.scatter(X1, X2, s=area, c=colors, alpha=0.5)
plt.title('test')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
#畫two-moons 資料集 (依 y值 1,0), 以mglearn.discrete_scatter()
X,y=make_moons(n_samples=100, noise=0.25, random_state=3)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.legend(["0,1"],loc=4)
plt.title('two-moons')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
