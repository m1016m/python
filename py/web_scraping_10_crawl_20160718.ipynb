```python
#a script that retrieves an arbitrary
#Wikipedia page and produces a list of links on that page:
```

```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup
```

```python
html=urlopen("http://en.wikipedia.org/wiki/Kevin_Bacon") 
```

```python
bsObj=BeautifulSoup(html,"lxml") 
```

```python
a_tag_list=bsObj.findAll("a") #find all "a" tag, return a list of "a" tag
```

```python
a_tag_list #list of tag <a>
```

```python
for l in a_tag_list:            #從 a_tag_list 迭代每個元素 l (tag <a>)
    if 'href' in l.attrs:       #如果 l 有 'href' 屬性 
        print(l.attrs['href'])  #印出 l 中 'href' 的值
```

```python
#練習: 將上例以 list comprehension 改寫
```

```python
link_list=[l.attrs['href'] for l in bsObj.findAll("a") if 'href' in l.attrs] 
```

```python
for ln in link_list:
    print(ln)
```
