```python
#convert each list of food nutrients to a DataFrame, add a column for the food id, and
#append the DataFrame to a list. Then, these can be concatenated together with concat():
```

```python
nutrients=[] #空串列 nutrients
```

```python
for rec in db: #db--> list of dict., 迭代每一個 dict.
    fnuts=DataFrame(rec['nutrients'])
    fnuts['id']=rec['id']
    nutrients.append(fnuts)
```

```python
nutrients=pd.concat(nutrients, ignore_index=True) #將list nutrients 中的元素一個一個接合
                                                  #pandas concat() 很像SQL union　
```

```python
nutrients
```

```python
#丟棄重複項次
```

```python
nutrients.duplicated().sum()
```

```python
nutrients=nutrients.drop_duplicates()
```

```python
info.keys()
```

```python
#為避免名稱重複,我們將更改欄位名稱。首先,寫一個 dict. "col_mapping"
col_mapping={'description':'food','group':'fgroup'}
```

```python
info=info.rename(columns=col_mapping,copy=False)
```

```python
info
```

```python
col_mapping={'description':'nutrients','group':'nutgroup'} #dict col_mapping
```

```python
nutrients=nutrients.rename(columns=col_mapping,copy=False)
```

```python
nutrients
```

```python
ndata=pd.merge(nutrients,info, on='id',how='outer') #以pd.merge() 作 outer join
```

```python
ndata
```

```python
ndata.ix[30000] #以DataFrame.ix[index] 
```

```python
ndata.head(3)
```

```python
#依 'nutrients,fgroup' 分類, 並求中位數 (quantile(0.5))
result=ndata.groupby(['nutrients','fgroup'])['value'].quantile(0.5)
```

```python
result #result, 具多重索引的 Series Object
```
