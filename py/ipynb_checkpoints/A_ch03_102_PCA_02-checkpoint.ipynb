```python
from sklearn.datasets import load_breast_cancer
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import mglearn
```

```python
#load cancer dataset
cancer=load_breast_cancer()
```

```python
#以 subplots() 建立 15x2 的畫布矩陣
fig,axes=plt.subplots(15,2,figsize=(10,20))

#malignant, ndarray shape (212,30), 癌症資料點
#benign, ndarray shape (357,30), 良性資料點 
malignant=cancer.data[cancer.target==0]
benign=cancer.data[cancer.target==1]

#以ravel將axes展平
ax=axes.ravel()

#
for i in range(30):
    _,bins=np.histogram(cancer.data[:,i],bins=50) #_,代表未使用的變數
    ax[i].hist(malignant[:,i],bins=bins,color=mglearn.cm3(0),alpha=.5) #alpha: 透明度
    ax[i].hist(benign[:,i],bins=bins,color=mglearn.cm3(2),alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(()) 
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Freauence")
ax[0].legend(["malignant","benign"],loc='best')
fig.tight_layout()
```

```python
#使用PCA找到前面兩個主成分，並在新座標下用散佈圖檢視這些資料。
```

```python
#step 0. 應用PCA之前，我們以StandardScaler做預處理，使每個特徵具單位標準差。
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler() #建立一個 StandardScaler Object 'scaler' 
scaler.fit(cancer.data) #以 cancer.data fit 'scalar' 
X_scaled=scaler.transform(cancer.data) #以 fit 完成的 'scaler' 
                                       #對 cancer.date 做 transform
```

```python
#step 1. PCA 轉換
```

```python
#建立一個PCA object 'pca', 保留前兩個主成分 (n_components=2)
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
#以StandardScaler轉換過的資料 'X_scaled'來訓練 pca
pca.fit(X_scaled)
```

```python
#以訓練完成的 'pca' 對 'X_scaled' 做轉換
#將資料轉換成前兩個主成分 'X_pca',
X_pca=pca.transform(X_scaled)
```

```python
#檢視PCA轉換前後資料的shape (X_scaled / X_pca)
print('X_scaled.shape: ',X_scaled.shape,' X_pca.shape: ',X_pca.shape)
```

```python
#以 mglearn 畫出前兩個主成分
```

```python
plt.figure(figsize=(8,8)) #建立一個 (8,8)的畫布
mglearn.discrete_scatter(X_pca[:,0],X_pca[:,1],cancer.target)
plt.legend(cancer.target_names,loc='best')
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

```python
#使用 LogisticRegression 分類器對第一主成分(X_pca[:,0])做分類
```

```python
from sklearn import linear_model
```

```python
#建立一個LogisticRegression分類器
logistic = linear_model.LogisticRegression() 
```

```python
#http://scikit-learn.org/stable/auto_examples/plot_digits_pipe.html#sphx-glr-auto-examples-plot-digits-pipe-py
```

```python
#原始資料 fit logistic model
#載入 train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
cancer.data,cancer.target,random_state=0)
```

```python
#原始資料 fit logistic model
logistic.fit(X_train,y_train)
```

```python
#Train score / Test score 
print('Train score:',logistic.score(X_train,y_train))
print('Test score:',logistic.score(X_test,y_test))
```

```python
#==============================================================
#使用PCA轉換後的第一,二主成分,以邏輯斯回歸分類器做分類
```

```python
#第一,二主成分 (X_pca), 使用 train_test_split()
X_train,X_test,y_train,y_test=train_test_split(
X_pca,cancer.target,random_state=0)
```

```python
#create another model 'logistic_pca2'
logistic_pca2 = linear_model.LogisticRegression() 
```

```python
#以第一,二主成分資料集 fit model 'logistic_pca2'
logistic_pca2.fit(X_train,y_train)
```

```python
#Train score / Test score 
print('Train score:',logistic_pca2.score(X_train,y_train))
print('Test score:',logistic_pca2.score(X_test,y_test))
```
