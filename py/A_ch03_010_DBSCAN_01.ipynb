```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import mglearn
```

```python
#import DBSCAN module
from sklearn.cluster import DBSCAN
```

```python
#建立一個小樣本(12)的虛擬資料集
from sklearn.datasets import make_blobs
X,y=make_blobs(random_state=0,n_samples=12)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
#以scikit-learn DBSCAN 模型預設值，將X分群
dbscan=DBSCAN()
clusters=dbscan.fit_predict(X)
```

```python
#檢視分群結果, 所有資料點都被視為 noise point (-1)
#因為eps和min_samples採預設值
clusters
```
